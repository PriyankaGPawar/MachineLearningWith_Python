```
#Public Scope
#Private Scope -> __  (double underscore)
#Protected Scope ->  _ (single underscore)
from textblob import TextBlob
from faker import Faker
import names
import random
fake = Faker()

class uber_data():
  # define a constrcutor or init function 
  def __init__(self):
    self.uber_database = {}  #Public 
    self.__uber_database_protected = {}  #Private
    self.month = []
  def add_user(self,name):
    # This function is just adding a user to database. this is always the first step 
    # Once the user is added - then you will be adding the properties 
    if name not in self.uber_database:
      self.uber_database[name] = {} # initially, in DB, you have to first add a user 
      self.__uber_database_protected[name] = {}
      self.uber_database[name]['Rating'] = {}
      self.uber_database[name]['Reviews'] = {}
      self.uber_database[name]['Sentiment'] = {}
      return "User Added in Data -",name
    else:
       return "User Already Exsist, please add new name -",name

  def add_rating(self,name,month,rating):
    if name in self.uber_database:
      self.uber_database[name]['Rating'][month] = rating
      return "Month & Rating is Added"
    else:
      return "User Dosn't Exsist, Run Add User Function First"

  def add_review(self,name,month,review):
       if name in self.uber_database:
        self.uber_database[name]['Reviews'][month] = review
        return "Review is Added"
       else:
        return "User Dosn't Exsist, Run Add User Function First"

  def add_sentimnt(self, name,month,review):
       if name in self.uber_database:
        self.uber_database[name]['Sentiment'][month] = self._generate_sentiment(review)
        return "Sentiment is Added"
       else:
        return "User Dosn't Exsist, Run Add User Function First"

  def _generate_review(self):
    review = fake.text()
    return review
  
  def _generate_sentiment(self,review):
    return round(TextBlob(review).sentiment.polarity,2)

  def __add_dummy_data(self,number,monthlist): #Private Function that floods the database with sample records
    self.month = monthlist
    # Add 100 users 
    for i in range(number):
      name = names.get_full_name()
        #Add user 
      self.add_user(name)
      for monthname in self.month :
        # Add Month and Rating for the user
        self.add_rating(name,monthname,round(random.uniform(2.0, 5.0),2))
        review = self._generate_review()
        self.add_review(name,monthname,review)
        self.add_sentimnt(name,monthname,review)

  def __export_data(self,name):
    file_open = open('ratings.txt',"w")
    file_open.write(str(self.uber_database[name]))
    file_open.close()
    return "File Saved with Data"
  
```

```
#Define an object of class
x = uber_data() # no parameter since your init function has no parameter
```
